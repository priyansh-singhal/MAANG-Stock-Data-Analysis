```python
import pandas as pd
import numpy as np
```

## Importing the required Data

```python
meta    = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/FB.csv')
amazon  = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/AMZN.csv')
apple   = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/AAPL.csv')
nvidia  = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/NVDA.csv')
google  = pd.read_csv('/kaggle/input/stock-market-dataset/stocks/GOOGL.csv')
```

```python
dataframes = [meta,amazon,apple,nvidia,google]
```

## Moving Average

```python
for df in dataframes:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
meta.head(60)
```

## Close

```python
for df in dataframes:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
apple.head()
```

```python
for df in dataframes:
    df["Change in price"] = df["Close"] - df["Previous day close price"]
```

```python
apple.head()
```

```python
for df in dataframes:
    df["Percentage change in price"] = df.Close.pct_change()
```

```python
apple.head()
```

```python
for df in dataframes:
    df['Previous day volumes'] = df.Volume.shift(1)
```

```python
apple.head()
```

```python
for df in dataframes:
    df["Change in volume"] = df['Volume'] - df["Previous day volumes"]
```

```python
apple.head()
```

```python
for df in dataframes:
    df["Percentage change in volume"] = df.Volume.pct_change()
```

## Save Dataframes as csv file

```python
meta.to_csv('meta.csv')
apple.to_csv('apple.csv')
amazon.to_csv('amazon.csv')
nvidia.to_csv('nvidia.csv')
google.to_csv('google.csv')
```
